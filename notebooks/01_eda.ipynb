### 2.2 Análise de Missing Values

**Por que é importante?**
- Missing values podem indicar problemas de coleta
- Modelos de ML precisam de estratégias para lidar com eles
- Alto % de missing pode indicar feature pouco confiável

```python
# Análise de missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).sort_values('Missing %', ascending=False)

print("ANÁLISE DE MISSING VALUES")
print("=" * 40)
if missing.sum() == 0:
    print("\nNenhum missing value encontrado!")
else:
    print(missing_df[missing_df['Missing Count'] > 0])
```

## 3. Análise da Variável Target (Default)

### 3.1 Distribuição das Classes

**Por que analisar?**
- Desbalanceamento afeta a escolha do modelo e métricas
- Credit scoring tipicamente tem classes desbalanceadas

```python
# Distribuição do target
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Contagem
target_counts = df['default'].value_counts()
colors = ['#2ecc71', '#e74c3c']
labels = ['Bom Pagador (0)', 'Mau Pagador (1)']

# Gráfico de barras
axes[0].bar(labels, target_counts.values, color=colors, edgecolor='black')
axes[0].set_title('Distribuição da Variável Target', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Contagem')
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 10, f'{v} ({v/len(df)*100:.1f}%)', ha='center', fontweight='bold')

# Pizza
axes[1].pie(target_counts.values, labels=labels, colors=colors, autopct='%1.1f%%',
            explode=(0, 0.05), shadow=True, startangle=90)
axes[1].set_title('Proporção de Classes', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('../reports/figures/eda_target_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

# Razão de desbalanceamento
ratio = target_counts[0] / target_counts[1]
print(f"\nRazão de desbalanceamento: {ratio:.2f}:1 (Bons:Maus)")
print(f"   Isso significa que para cada mau pagador, temos {ratio:.1f} bons pagadores.")
```

### 3.2 Implicações do Desbalanceamento

Com ~70% de bons pagadores vs ~30% de maus pagadores:

1. **Acurácia não é boa métrica** - Um modelo "burro" que prevê sempre "bom" teria 70% acurácia
2. **Precisamos de métricas balanceadas**: ROC-AUC, F1, Precision-Recall
3. **Estratégias de balanceamento**: `class_weight='balanced'` ou `scale_pos_weight`

## 6. Análise de Correlação

### 6.1 Correlação entre Features Numéricas

**Por que é importante?**
- Features muito correlacionadas = **multicolinearidade** 
- Prejudica interpretação de modelos lineares
- Pode indicar features redundantes

```python
# Matriz de correlação
corr_matrix = df[numerical_cols + ['default']].corr()

# Heatmap
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Triângulo superior
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, vmin=-1, vmax=1, square=True, linewidths=0.5)
plt.title('Matriz de Correlação - Features Numéricas', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/eda_correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Correlação com o target (ordenada)
target_corr = corr_matrix['default'].drop('default').sort_values(key=abs, ascending=False)

print("CORRELAÇÃO COM TARGET (default):")
print("=" * 40)
for feat, corr in target_corr.items():
    direction = "↑" if corr > 0 else "↓"
    print(f"{feat:20s}: {corr:+.3f} {direction}")
```

### 6.2 Correlação de Features Categóricas com Target

Usamos **Cramér's V** para medir associação entre variáveis categóricas.

```python
def cramers_v(x, y):
    """Calcular Cramér's V para associação entre variáveis categóricas."""
    contingency = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(contingency)[0]
    n = len(x)
    min_dim = min(contingency.shape) - 1
    return np.sqrt(chi2 / (n * min_dim)) if min_dim > 0 else 0

# Calcular Cramér's V para cada feature categórica
cramers_scores = {}
for col in categorical_cols:
    cramers_scores[col] = cramers_v(df[col], df['default'])

cramers_df = pd.DataFrame.from_dict(cramers_scores, orient='index', columns=["Cramér's V"])
cramers_df = cramers_df.sort_values("Cramér's V", ascending=False)

print("ASSOCIAÇÃO CATEGÓRICA COM TARGET (Cramér's V):")
print("=" * 50)
print("Interpretação: 0 = nenhuma associação, 1 = associação perfeita")
print("                > 0.1 = fraca, > 0.3 = moderada, > 0.5 = forte")
print("-" * 50)
print(cramers_df.round(3).to_string())
```

## 7. Feature Selection - Critérios de Seleção

### 7.1 Ranking de Features por Importância

**Critérios utilizados:**
1. **Correlação com target** (numéricas) - quanto maior, melhor
2. **Cramér's V** (categóricas) - quanto maior, melhor
3. **Variância** - features constantes são inúteis
4. **Taxa de missing** - alto missing = pouca confiabilidade

```python
# Criar DataFrame de importância
importance_data = []

# Features numéricas
for col in numerical_cols:
    importance_data.append({
        'feature': col,
        'type': 'numerical',
        'metric': 'Correlation',
        'value': abs(corr_matrix.loc[col, 'default']),
        'variance': df[col].var(),
        'missing_pct': df[col].isnull().mean() * 100
    })

# Features categóricas
for col in categorical_cols:
    importance_data.append({
        'feature': col,
        'type': 'categorical',
        'metric': "Cramér's V",
        'value': cramers_scores[col],
        'variance': df[col].nunique(),  # n_categories como proxy
        'missing_pct': df[col].isnull().mean() * 100
    })

importance_df = pd.DataFrame(importance_data)
importance_df = importance_df.sort_values('value', ascending=False)

# Visualizar
plt.figure(figsize=(12, 8))
colors = ['#3498db' if t == 'numerical' else '#e74c3c' for t in importance_df['type']]
bars = plt.barh(range(len(importance_df)), importance_df['value'], color=colors)
plt.yticks(range(len(importance_df)), importance_df['feature'])
plt.xlabel('Importância (Correlação / Cramér\'s V)')
plt.title('Ranking de Features por Importância', fontsize=14, fontweight='bold')

# Legenda
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#3498db', label='Numérica'),
                   Patch(facecolor='#e74c3c', label='Categórica')]
plt.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.savefig('../reports/figures/eda_feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()
```

### 7.2 Decisões de Feature Selection

Com base na análise:

```python
# Top features por tipo
print("=" * 60)
print("RECOMENDAÇÕES DE FEATURE SELECTION")
print("=" * 60)

# Threshold para seleção
threshold = 0.05

selected = importance_df[importance_df['value'] >= threshold]
dropped = importance_df[importance_df['value'] < threshold]

print(f"\nFEATURES SELECIONADAS ({len(selected)})")
print("-" * 40)
for _, row in selected.iterrows():
    print(f"  {row['feature']:20s} ({row['type']:11s}): {row['value']:.3f}")

if len(dropped) > 0:
    print(f"\nFEATURES COM BAIXA IMPORTÂNCIA ({len(dropped)})")
    print("-" * 40)
    for _, row in dropped.iterrows():
        print(f"  {row['feature']:20s} ({row['type']:11s}): {row['value']:.3f}")

print("\n" + "=" * 60)
print("NOTA: Para este projeto, mantivemos TODAS as features.")
print("Em produção, considere remover features com importância < 0.05")
print("=" * 60)
```
